# Sentiment Analysis Notebook

This notebook demonstrates how to run `sentiment analysis.py` from Python, classify a sample tweet, process tweets from a CSV file, and save the classification results.

## Steps

1. Install dependencies: `nltk` and optionally `pandas`
2. Load the sentiment analysis script
3. Ensure the VADER lexicon is available
4. Classify sample text and CSV data
5. Save results to a new CSV file

> If `pandas` is not installed, the notebook will still run using Python's built-in CSV reader.


```python
import csv
import os
import sys
from pathlib import Path
import importlib.util

try:
    import pandas as pd
except ImportError:
    pd = None

import nltk

nb_dir = Path.cwd()
script_path = nb_dir / 'sentiment analysis.py'
print('Loading sentiment analysis script from:', script_path)

spec = importlib.util.spec_from_file_location('sentiment_analysis', str(script_path))
sentiment_analysis = importlib.util.module_from_spec(spec)
spec.loader.exec_module(sentiment_analysis)

ensure_vader = sentiment_analysis.ensure_vader
classify_sentiment = sentiment_analysis.classify_sentiment
classify_tweet_lines = sentiment_analysis.classify_tweet_lines
read_csv_tweets = sentiment_analysis.read_csv_tweets
save_output = sentiment_analysis.save_output
SentimentIntensityAnalyzer = sentiment_analysis.SentimentIntensityAnalyzer
```

```python
ensure_vader()
print('VADER lexicon is available and ready to use.')
```

```python
analyzer = SentimentIntensityAnalyzer()
sample_tweet = 'I love the new update!'
result = classify_sentiment(sample_tweet, analyzer)
print('Sample tweet result:')
print(f"Label: {result['label']}")
print(f"Compound score: {result['compound']:+.3f}")
print(f"Text: {result['text']}")
```

```python
csv_path = os.path.abspath('sample_tweets.csv')
with open(csv_path, 'w', encoding='utf-8', newline='') as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(['text'])
    writer.writerow(['I love this product!'])
    writer.writerow(['This is the worst service ever.'])
    writer.writerow(['I guess it is okay.'])

print('Sample CSV created at:', csv_path)

tweets = read_csv_tweets(csv_path)
results = classify_tweet_lines(tweets, analyzer)
print('\nCSV classification results:')
for item in results:
    print(f"{item['label'].upper():>8} | {item['compound']:+.3f} | {item['text']}")
```

```python
output_csv = os.path.abspath('sentiment_results.csv')
save_output(results, output_csv)
print(f'Classification results saved to: {output_csv}')
```
